```python
st2 = pd.read_excel('Station_2.xlsx')
st1 = st2.drop(columns = ['Unnamed: 0'])
coor = pd.read_excel('ชื่อสถานีและพิกัด.xlsx')
coor.loc[coor['สถานี']==171181]
```

```python
station = coor.iloc[:37, 1:9]
station.rename(columns = {'สถานี' : 'Station'}, inplace = True)
station
```

```python
import pyproj
from functools import partial
from pyproj import Proj, transform
```

```python
proj_wgs84 = Proj(init = "epsg:24047")
proj_gk4 = Proj(init = "epsg:4326")

transformer = partial(transform, proj_wgs84, proj_gk4)
lon, lat = transformer (station['E'].values, station['N'].values)
lon, lat
```

```python
st_coor = {'Latitude' : lat, 'Lontitude' : lon, }
coordinate = pd.DataFrame(st_coor)
coordinate.shape
```

```python
station.shape
```

```python
station_coordinate = pd.concat([station, coordinate],axis=1)
station_coordinate = station_coordinate.drop(columns = ['Unnamed: 2', 'ที่ตั้ง', 'จังหวัด' , 'Lat' , 'Lon'])
station_coordinate
```

```python
station_coordinate.to_excel('station_coordinate.xlsx')
```

```python
station_coordinate.shape
```

```python
import folium
station_map = folium.Map(location = [18.5,98.5], zoom_start = 8.5)
locate = {}

for i, j, k in zip(station_coordinate['Station'], station_coordinate['Latitude'], station_coordinate['Lontitude']) :
    temp = []
    temp.extend((j,k))
    locate ['loc'] = temp
    marker = folium.Marker(location = locate ['loc'], popup = str (k) + 'Type: ' + str (j))
    marker.add_to(station_map)
    
station_map
```
